## 1.How many are there in each Team and the percentage splitting with respect to the total employees?

```python
df['Team'].value_counts()
```

```python
total_emp=df['Team'].value_counts()
#split w.r.t the total employees
percentage_split=total_emp/len(df)*100

team_info=pd.DataFrame({'Total employee':total_emp,'percentage split':percentage_split})
print(team_info)
```

```python
import matplotlib.pyplot as plt
plt.bar(total_emp.index,total_emp.values,color='blue')
plt.xlabel('Team')
plt.title('percentage splitting of employees w.r.t each team')
plt.xticks(rotation='vertical')
plt.show()
```

__It is clear that Team "New Orleans Pelicans" has the highest percentage of employees belongs to.other Team have almost the same percentage.Consequently,the splitting percentage, or the proportion of the total workforce,is also highest for these respective Team.The bar plot is used to plot the same data which illustrates the percentage splitting with respect to the employees__

## 2.Seggregate the employees w.r.t different position?

```python
df['Position'].value_counts()
```

```python
df[['Position','Name']]
```

```python
#seggregating employees w.r.t different position
#display with only name and position

grouped_by_position=df.groupby('Position')
for position,group in grouped_by_position:
    print(f"Position:{position}")
    print(group[['Name','Position']])
    print("\n")
```

```python
#plot the data using countplot
unique_positions=df['Position'].unique()
print("unique positions:",unique_positions)
plt.figure(figsize=(7,5))
plt.title('seggregate Employee by position')
sns.countplot(x='Position',data=df,hue='Position',palette='dark')
```

__By analysing the above countplot,it is clear that,the most number of employees are belong to 'SG' position,and least number of employees belong to 'C' position.other position may has different counts,but 'SG' has the highest count followed closely by 'PF'__

## 4.Find out under which team and position,spending in terms of salary is high?

```python
x=df['Salary'].mean()
df['Salary'].fillna(x,inplace=True)
df
```

```python
data=df[['Team','Position','Salary']]
data
```

```python
salary=data.groupby(['Team','Position'])['Salary'].sum()
max_spending=salary.idxmax()
high_salary=salary.max()
print(f"The Team and position with highest spending=\n Team:{max_spending[0]}\n position:{max_spending[1]}\n high salary= {high_salary}")
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
plt.figure(figsize=(15,5))
sns.boxplot(x='Team',y='Salary',hue='Position',data=df)
plt.xlabel('Team')
plt.ylabel('Salary')
plt.title('spending in terms of high salary')
plt.xticks(rotation='vertical')
plt.show()
```

__The Los Angeles Lakers spend the highest in terms of salary and it is about 31866445 with the position 'SF'.The above boxplot illustrates the salary distribution across different positions within each team in the dataset.it shows the distribution of quantitative data in a way that facilitates comparisons between variables or across levels of categorical variable.__

## 5.Find if there is any correlation between age and salary,represent it visually.

```python
df1=df[['Age','Salary']]
df1
```

```python
corr=df1.corr()
print(corr)
plt.figure(figsize=(7,6))
sns.heatmap(df[['Age','Salary']].corr(),annot=True,cmap='Greens')
plt.title('Correlation between Age and salary')
plt.show()
```

__The correlation shows the relationship between "Age" and 'Salary'.Here heatmap is used to plot the data that shows the magnitude of a certain phenomenon in the form of colors.cmap is the list of colors,optional.The value of 1.00000 along the diagonal represents the correlation of a variable with itself,which is always 1.it indicates positive correlaton.The correlation of 0.214009 is positive but weak relationship between age and salary.By analysing the above heatmap we can conclude that the employees in the age group of 30-40 years have the highest salary.It is showing that as the age increases the salary also increases.__
